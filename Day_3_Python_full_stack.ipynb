#Amstrong number or not

```python
n = int(input("Enter number: "))
digits = len(str(n))
total = sum(int(d)**digits for d in str(n))

if total == n:
    print("Armstrong")
else:
    print("Not Armstrong")
```

#perfect number or not

```python
n = int(input("Enter a number: "))
s = sum(i for i in range(1, n) if n % i == 0)

if s == n:
    print("Perfect number")
else:
    print("Not a perfect number")
```

#Prime number or not

```python
n = int(input("Enter a number: "
))

if n < 2:
    print("Not a prime number")
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

#Check whether leap year or not

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year ")
else:
    print("Not a leap year ")
```

#Simple calculator

```python
a = float(input("Enter first number: "))
op = input("Enter operator (+, -, *, /): ")
b = float(input("Enter second number: "))

if op == "+":
    print("Result:", a + b)
elif op == "-":
    print("Result:", a - b)
elif op == "*":
    print("Result:", a * b)
elif op == "/":
    if b == 0:
        print("Division by zero not allowed ")
    else:
        print("Result:", a / b)
else:
    print("Invalid operator ")
```

# Using while loop count no.of even digits in  number

```python
n = int(input("Enter a number: "))
temp = n
count = 0

while temp > 0:
    digit = temp % 10

    if digit % 2 == 0:
        count += 1
    temp = temp // 10

print("Number of even digits:", count)
```

#Check if a string contains only numbers

```python
s = input("Enter a string: ")

flag = 1

for i in s:
    if i < '0' or i > '9':
        flag = 0

if flag == 1:
    print("Only Numbers")
else:
    print("Not Only Numbers")
```

#Find the longest word in the sentence

```python
s = input("Enter a sentence: ")

words = s.split()

longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```

#Reverse the words in the sentence

```python
s = input("Enter a sentence: ")

words = s.split()

for i in range(len(words)-1, -1, -1):
    print(words[i], end=" ")
```

#Counting vowels and consonents

```python
s = "fghy57687ijh hy6"

vo = 0
con = 0
dig = 0
space = 0

for ch in s:
    if ch in "aeiouAEIOU":
        vo += 1
    elif ch.isdigit():
        dig += 1
    elif ch == " ":
        space += 1
    else:
        con += 1

print("Vowels:", vo)
print("Consonants:", con)
print("Digits:", dig)
print("Spaces:", space)
```

#Counting repeated letters

```python
word = "Data Analyst"
freq = {}

for ch in word:
    if ch != " ":
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

for key in freq:
    if freq[key] > 1:
        print(key, "=", freq[key])
```

#Palindrome

```python
s = "A place, paris, germany"
cleaned = s.replace(",", "").replace(" ", "").lower()
print(cleaned)
s1 = cleaned[::-1]
if cleaned == s1:
    print("Palindrome")
else:
    print("Not a palindrome")
```
