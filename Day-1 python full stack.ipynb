```python
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
print("\n")
```

```python

x = 1
y = 11
z=int(input("Enter a number: "))

while (x < y):
    print(z, "x", x, "=", z * x)
    x += 1
print("\n")
```

# days of week selection using if elif else

```python

n = int(input("Enter a number to get a day: "))
if n == 1:
    print("Monday")
elif n == 2:
    print("Tuesday")
elif n == 3:
    print("Wednesday")
elif n == 4:
    print("Thursday")
elif n == 5:
    print("Friday")
elif n == 6:
    print("Saturday")
elif n == 7:
    print("Sunday")
else:
    print("Invalid")
print("\n")
```

# college grading system

```python

marks = int(input("Enter Marks to get grade: "))
if marks > 90:
    print("S grade")
elif marks <= 90 and marks >= 80:
    print("A grade")
elif marks < 80 and marks >= 70:
    print("B grade")
elif marks < 70 and marks >= 60:
    print("C grade")
elif marks < 60 and marks >= 50:
    print("D grade")
elif marks < 50 and marks >= 40:
    print("E grade")
else:
    print("Fail / Invalid input")
print("\n")
```

# even odd separate

```python

n = int(input("Enter the number to separate odds and evens: "))
print("Even numbers:")
even = []
for i in range(1, n + 1):
    if i % 2 == 0:
        even.append(i)
print(even)

print("Odd numbers:")
odd = []
for i in range(1, n + 1):
    if i % 2 != 0:
        odd.append(i)
print(odd)
```

#sum of digits

```python

f = int(input("Enter number to make sum of digits: "))
tot = 0
while f > 0:
    digit = f % 10
    tot += digit
    f = f // 10
print("Sum of digits:", tot)
```

#prod of digits

```python

f = int(input("Enter number to make prod of digits: "))
prod = 1
while f > 0:
    digit = f % 10
    prod = prod * digit
    f = f // 10
print("Prod of digits:", prod)
```

# reverse of number for 3 digit only

```python
# reverse of number for 3 digit only
num = int(input("Enter a number to reverse: "))
original_num = num

ones = num % 10
tens = (num // 10) % 10
hundreds = num // 100

rev = (ones * 100) + (tens * 10) + hundreds
print("Reversed number:", rev)

if rev == original_num:
    print("Palindrome")
else:
    print("Not palindrome")
```

#reverse a number for any digit number

```python
# reverse a number for any digit number
num = int(input("Enter a number to reverse: "))
original_num = num
rev = 0

while num > 0:
    b = num % 10
    rev = (rev * 10) + b
    num = num // 10

print("Reversed number:", rev)
if rev == original_num:
    print("Palindrome")
else:
    print("Not palindrome")
```

# perfect number or not

```python
# perfect number or not
perf = int(input("Enter a number: "))
lst = []

for i in range(1, perf):
    if perf % i == 0:
        lst.append(i)
print(lst)

added = sum(lst)
if added == perf:
    print("Perfect number")
else:
    print("Not perfect number")
```
