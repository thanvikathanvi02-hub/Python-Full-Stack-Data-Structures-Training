use conditional statements to find the greatest number

```python
a = int(input("Enter 1st number: "))
b = int(input("Enter 2nd number: "))
c = int(input("Enter 3rd number: "))
d = int(input("Enter 4th number: "))
e = int(input("Enter 5th number: "))

greatest = a

if b > greatest:
    greatest = b
if c > greatest:
    greatest = c
if d > greatest:
    greatest = d
if e > greatest:
    greatest = e

print("Greatest number is:", greatest)
```

Hollow Square pattern

```python
n = 5

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

Hollow pyramid

```python
n = 5

for i in range(1, n + 1):
    for j in range(1, 2 * n):
        if i == n or j == n - i + 1 or j == n + i - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

Hollow inverted pyramid

```python
n = 5

for i in range(n, 0, -1):
    for j in range(1, 2 * n):
        if i == n or j == n - i + 1 or j == n + i - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

Hollow Diamond pattern

```python
n = 5


for i in range(1, n + 1):
    for j in range(1, 2 * n):
        if j == n - i + 1 or j == n + i - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

for i in range(n - 1, 0, -1):
    for j in range(1, 2 * n):
        if j == n - i + 1 or j == n + i - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

#Floyd’s Triangle

```python
n = 5
num = 1

for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

Binary triangle

```python
n = 5

for i in range(1, n + 1):
    for j in range(i):
        if (i + j) % 2 == 0:
            print("1", end=" ")
        else:
            print("0", end=" ")
    print()
```

Hollow hexagon

```python
n = 6

for i in range(n):
    print(" " * (n - i) + "* " * (n + i))

for i in range(n):
    print("* " * (2 * n))

for i in range(n - 1, -1, -1):
    print(" " * (n - i) + "* " * (n + i))
```

Hollow rectangle

```python
rows = 5
cols = 8

for i in range(rows):
    for j in range(cols):
        if i == 0 or i == rows-1 or j == 0 or j == cols-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

ATM PROGRAAM

```python
#1. Available balance 2. Deposit 3. Withdraw
#Bank account 11 digit number
#password SRGC$2026-1

code = "SRGC$2026-1"
availbal = 100

accno = list(input("Enter the bank acc number 11 digits: "))

if len(accno) == 11:
    print("Enter Password: ")
else:
    print("Invalid")

password = input()

if password == code:
    print("Welcome!")
else:
    print("Wrong password")

print("Enter 1 for balance check 2 for deposit 3 for withdraw")
choice = int(input())

if choice == 1:
    print("Available Balance: ", availbal)

elif choice == 2:
    print("Enter amount to deposit: ")
    amt = int(input())
    availbal = availbal + amt
    print("New Balance: ", availbal)

elif choice == 3:
    print("Enter amount to withdraw: ")
    amt = int(input())
    availbal = availbal - amt
    print("New Balance: ", availbal)

else:
    print("Try Again")
```

#Greatest of All Numbers

```python
arr = list(map(int, input("Enter numbers: ").split()))
for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] > arr[j]:
            arr[i], arr[j] = arr[j], arr[i]
print(arr[-2])
```
