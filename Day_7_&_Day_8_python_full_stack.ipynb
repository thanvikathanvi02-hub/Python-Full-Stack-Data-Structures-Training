#Sum of Digits Using Recursion

```python
def sumofdigits(n):
    if n == 0:
        return 0
    return (n % 10) + sumofdigits(n // 10)

print(sumofdigits(12345))
```

#Reverse a Number Using Recursion

```python
def palindrome(m):
    if m < 10:
        return m
    return (m % 10) * 10 + palindrome(m // 10)

print(palindrome(121))
```

#Check Palindrome String Using Recursion

```python
def palindrome(s):
    if len(s) < 1:
        return True
    if s[0] != s[-1]:
        return False
    return palindrome(s[1:-1])

print(palindrome("madam"))
```

#Find Power of a Number

```python
def power(base, power):
    if power == 0:
        return 1
    return base ** power

print(power(2, 3))
```

#Power Using Recursion

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    return base * power(base, exponent - 1)

print(power(2, 3))
```

#Valid Parentheses Using Stack

```python
class Solution:
    def isValid(self, s):
        stack = []

        for ch in s:
            if ch == '(' or ch == '{' or ch == '[':
                stack.append(ch)
            else:
                if not stack:
                    return False

                top = stack.pop()

                if ch == ')' and top != '(':
                    return False
                if ch == '}' and top != '{':
                    return False
                if ch == ']' and top != '[':
                    return False

        return not stack

s = Solution()
print(s.isValid("{[}"))
```

#Min Stack Implementation

```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, value):
        self.stack.append(value)

        if not self.min_stack or value <= self.min_stack[-1]:
            self.min_stack.append(value)

    def pop(self):
        if self.stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()

        self.stack.pop()

    def top(self):
        return self.stack[-1]

    def getMin(self):
        return self.min_stack[-1]
```

#Queue Using Two Stacks

```python
class MyQueueusingstack:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def push(self, x):
        self.stack1.append(x)

    def pop(self):
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())

        return self.stack2.pop()

    def peek(self):
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())

        return self.stack2[-1]

    def empty(self):
        return not self.stack1 and not self.stack2
```

#Count Ways to Climb Stairs (1,2,3 Steps)

```python
def Steps(n):
    if n < 3:
        return n
    else:
        return Steps(n-1) + Steps(n-2) + Steps(n-3)

print(Steps(5))
```

#Find Nth Term of Arithmetic Progression Using Recursion

```python
def AP(a, d, n):
    if n == 1:
        return a
    else:
        return AP(a, d, n-1) + d

print(AP(2, 1, 3))
```
