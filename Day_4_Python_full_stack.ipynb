#Count Frequency of Elements

```python
nums = [2,2,1,4,5,1,6,6,6,4]

snums = sorted(nums)

count = {}

for num in snums:
    if num not in count:
        count[num] = 1
    else:
        count[num] += 1

print(count)
```

#Three Sum

```python
class Solution:
    def threeSum(self, nums, target):

        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                for k in range(j+1,len(nums)):

                    if nums[i] + nums[j] + nums[k] == target:
                        return [i,j,k]
```

#Two Sum

```python
nums = [2,7,11,15]
target = 9

seen = {}

for i,num in enumerate(nums):

    complement = target - num

    if complement in seen:
        print(seen[complement],i)
        break

    seen[num] = i
```

#Merge Two Sorted Arrays

```python
a = [2,5,8,9]
b = [3,4,7,8]

c = [0] * (len(a)+len(b))

i = j = k = 0

while i < len(a) and j < len(b):

    if a[i] <= b[j]:
        c[k] = a[i]
        i += 1
    else:
        c[k] = b[j]
        j += 1

    k += 1

while i < len(a):
    c[k] = a[i]
    i += 1
    k += 1

while j < len(b):
    c[k] = b[j]
    j += 1
    k += 1

print(c)
```

#Union of Two Sorted Arrays

```python
a = [2,5,8,9]
b = [3,4,7,8]

c = [0] * (len(a)+len(b))

i = j = k = 0

while i < len(a) and j < len(b):

    if a[i] < b[j]:
        c[k] = a[i]
        i += 1

    elif a[i] > b[j]:
        c[k] = b[j]
        j += 1

    else:
        c[k] = a[i]
        i += 1
        j += 1

    k += 1

print(c[:k])
```

#Intersection of Two Arrays

```python
a = [2,5,8,9]
b = [3,4,7,8]

visited = [0] * len(b)

c = [0] * min(len(a),len(b))

k = 0

for i in range(len(a)):
    for j in range(len(b)):

        if a[i] == b[j] and visited[j] == 0:
            c[k] = a[i]
            k += 1
            visited[j] = 1
            break

result = c[:k]

print(result)
```
