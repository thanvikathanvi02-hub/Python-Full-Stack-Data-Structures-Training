# prime in a range(m,n) - Primes in a Range (returns list)
python

```python
def primeinarange(m, n):
    a = []
    for i in range(m, n+1):
        if i > 1:
            for j in range(2, i):
                if i % j == 0:
                    break
            else:
                a.append(i)
    return a

print(sum(primeinarange(2, 10)))
print(primeinarange(2, 10))
```

#even odd in a range(a,b) - Even and Odd lists in a Range

```python
def evenoddinarange(a,b):
    even = []
    odd = []
    for i in range(a, b+1):
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
    return even, odd

print("Even List:")
print(evenoddinarange(1, 30)[0])
print("Odd List:")
print(evenoddinarange(1, 30)[1])
```

#has pair(arr,t) - Two Sum / Has Pair with given sum

```python
def haspair(arr,t):
    left=0
    right=len(arr)-1
    while left<right:
        s=arr[left]+arr[right]
        if s==t:
            return True
        elif s<t:
            left+=1
        else:
            right-=1
    return False
print(haspair([1,2,3,4,5,6],10))
```

#quad eq(a,b,c) - Quadratic Equation Discriminant

```python
def quadeq(a,b,c):
    a = int(input("Enter a num: "))
    b = int(input("Enter a num: "))
    c = int(input("Enter a num: "))
    d = (b**2) - (4*a*c)
    if d > 0:
        print("Roots are real and unequal")
    elif d == 0:
        print("Roots are real and equal")
    else:
        print("Roots are imaginary")
    return d

print(quadeq(5,6,7))
```

# BinarySearch(a,high,low,target) - Binary Search


```python
def BinarySearch(a,high,low,target):
    while low <= high:
        mid = (high + low) // 2
        if a[mid] == target:
            return mid
        elif a[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
        return -1
a = list(map(int, input().split()))
target = int(input())
print(BinarySearch(a,len(a)-1,0,target))
```

#Solution.search - Search in Rotated Sorted Array

```python
class Solution(object):
    def search(self, nums, target):
        left = 0
        right = len(nums) - 1
        while left <= right:
            mid = (left + right)//2
            if nums[mid] == target:
                return mid
            if nums[left] <= nums[mid]:
                if nums[left] <= target < nums[mid]:
                    right = mid - 1
                else:
                    left = mid + 1
            else:
                if nums[mid] < target <= nums[right]:
                    left = mid + 1
                else:
                    right = mid - 1
        return -1

sol = Solution()
nums = list(map(int, input().split()))
target = int(input())
print(sol.search(nums, target))
```

#alphabet(n) - Vowels and Consonants

# Peak Element (inline)

```python
b = list(map(int, input().split()))
a = [0] + b + [0]

peaks = []

for i in range(1, len(a) - 1):
    if a[i] > a[i - 1] and a[i] > a[i + 1]:
        peaks.append(i - 1)

if len(peaks) == 0:
    print("No peaks")
else:
    largest = peaks[0]
    for i in peaks:

        if i > largest:
            largest = i
    print(largest)
```

#Solution.findPeakElement - Find Peak Element (Binary Search)

```python
class Solution(object):
    def findPeakElement(self,nums):
        left = 0
        right = len(nums) - 1
        while left < right:
            mid = (left + right) // 2
            if nums[mid] > nums[mid + 1]:
                left = mid + 1
            else:
                right = mid
        return left
```
