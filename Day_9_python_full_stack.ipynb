#Construct a linked list containing 5 nodes and set the head to node 3

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

node1 = Node(10)
node2 = Node(20)
node3 = Node(30)
node4 = Node(40)
node5 = Node(50)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

head = node3

print("Linked List:")
temp = node1
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next
print("NULL")

print("Head =", head.data)
```

#Insertion of a Node (at Beginning)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

print("Original List:")
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next
print("NULL")

newNode = Node(5)
newNode.next = head
head = newNode

print("After Insertion:")
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next
print("NULL")

head.next = head.next.next

print("After Deletion:")
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next
print("NULL")
```

#Insert After a Node

```python
class linkedlist():
    def __init__(self, data):
        self.data = data
        self.next = None

    def insert(self, data):
        new_node = linkedlist(data)
        new_node.next = self.next
        self.next = new_node

n1 = linkedlist(10)
n2 = linkedlist(100)
n3 = linkedlist(1000)

n1.insert(99)

n1.next.next = n2
n2.next = n3

head = n1

while head != None:
    print(head.data, end="->")
    head = head.next
print("None")
```

#Insert at Head

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def insert_at_head(head, data):
    new_node = Node(data)
    new_node.next = head
    return new_node

n1 = Node(10)
n3 = Node(40)
n4 = Node(5)

n1.next = n3
n3.next = n4

head = n1
head = insert_at_head(head, 99)

while head != None:
    print(head.data, end="->")
    head = head.next
print("None")
```

#Insert at Tail

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def insert_at_tail(head, data):
    new_node = Node(data)

    if head == None:
        return new_node

    temp = head
    while temp.next != None:
        temp = temp.next

    temp.next = new_node
    return head

n1 = Node(10)
n3 = Node(40)
n4 = Node(5)

n1.next = n3
n3.next = n4

head = n1
head = insert_at_tail(head, 99)

while head != None:
    print(head.data, end="->")
    head = head.next
print("None")
```

#Insert at Specific Position

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def insert_at_head(head, data):
    new_node = Node(data)
    new_node.next = head
    return new_node

def insert_at_pos(head, data, pos):
    if pos == 0:
        return insert_at_head(head, data)

    new_node = Node(data)
    temp = head

    for i in range(pos - 1):
        temp = temp.next

    new_node.next = temp.next
    temp.next = new_node

    return head
```

#Reverse a Linked List

```python
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

class Solution:
    def reverseList(self, head):
        prev = None
        curr = head

        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt

        return prev

n1 = ListNode(10)
n2 = ListNode(20)
n3 = ListNode(30)

n1.next = n2
n2.next = n3

head = Solution().reverseList(n1)

while head:
    print(head.val, end=" ")
    head = head.next
```
